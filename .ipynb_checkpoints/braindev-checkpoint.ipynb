# Modelling developmental trajectories of newborn brains

This notebook is about modelling developmental trajectories based on :

- One cross-sectional measurement per baby (post-birth)
- The knowledge of time of birth
- The knowledge of time of scan (i.e. when the measurement is taken)

We use the following piecewise linear model:

$$ y(t)= \begin{cases}c_{0}+\beta_{0} t & \text { if } t \leqslant t_{birth} \\ c_{1}+\beta_{1} t & \text { if } t \geqslant t_{birth}\end{cases} $$

Using continuity at birth, we have $y(t_{birth}) = c_0 + \beta_0 t_{birth} = c_1 + \beta_1 t_{birth} $.

A better parameterisation is in term of a time of onset ($t_{onset}$), pre-birth slope ($\beta_{pre}=\beta_0$) and post-birth slope ($\beta_{post} = \beta_1$). This will allow us, e.g., to set bounds on $t_{onset}$ to constrain the fitting. 

Let's now write the model for $y(t_{scan})$ as a function of the 3 parameters of interest:

$$ y(t_{scan}) = -\beta_0 * t_{onset} + \beta_0 * t_{birth} + \beta_1 * (t_{scan}-t_{birth}) $$.

One can immediately see that this model is in principle invertible if $t_{scan}$ and $t_{birth}$ are reasonably decorrelated, which we shall check.

```python

# choose a measure to be modelled
meas = 'csf_vol'

# Quick fit a model and look at the resuts
model   = 3
fm      = ForwardModel(model)
colours = ['k', 'g', 'b', 'r']
names   = fm.labels[:-1]

# Prepare the data
data,birth,scan = prepare(df,name=meas,deconfound=False,normalise=False)
# Do the fitting
samples, ML = do_fit(data,birth,scan,fm)

[sns.kdeplot(samples[:,i], color=colours[i], label=name) for i,name in enumerate(names)]
plt.title(meas)
plt.legend()

```

```python
# Plots

sns.set(style='whitegrid',font_scale=3)

model  = 1
fm     = ForwardModel(model)
meas   = 'gm_vol'
data,t_birth,t_scan = prepare(df,name=meas,deconfound=False,normalise=False)
samples, _ = do_fit(data,t_birth,t_scan,fm)
params = np.mean(samples,axis=0)

# Validation plots
beta0_term,beta1_term,_     = fm.get_params(params=params,group='term')
beta0_prem,beta1_prem,onset = fm.get_params(params=params,group='prem')


plt.figure(figsize=(45,15))

# BETA0 Validation1 : PLOT NEAR BIRTH DATA AGAINST AGE AT SCAN
# This should be similar to beta0

plt.subplot(1,3,1)
near = df.loc[(df['age_deltas']<1)] # babies scanned <1week since birth
sns.regplot(near['age_scan'],near[meas],'.',label='near birth data')
# model prediction
x = np.array([30,44])
y = near[meas].mean() + (x-near['age_scan'].mean())*beta0_term
plt.plot(x,y,'-',label='beta0 from model')
plt.legend()


# BETA1 : CHECK SLOPE AGAINST LONGITUDINAL SLOPES
long1,long2 = extract_long(df)
y1,y2 = np.array(long1[meas]),np.array(long2[meas])
t1,t2 = np.array(long1['age_deltas']),np.array(long2['age_deltas'])
# need to regress out age at birth
# y1 = regress_out(y1,long1[['age_birth']])
# y2 = regress_out(y2,long2[['age_birth']])
plt.subplot(1,3,2)
h1=plt.plot([t1,t2],[y1,y2],'ko-',alpha=.1,label='longitudinal data')
# model prediction
x = np.array([0,12])
y = beta0_prem*(long1['age_birth'].mean()-onset)+x*beta1_prem
h2 = plt.plot(x,y,label='beta1 from model')
plt.legend(handles=[h1[0],h2[0]])
plt.xlabel('age delta (weeks)')
```
